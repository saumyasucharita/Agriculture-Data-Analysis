```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
country = 'India'
```

```python
prod_ind_data = pd.read_csv('Dataset/archive/fao_data_production_indices_data.csv') #country_or_area
agri_data_all = prod_ind_data[prod_ind_data['category']=='agriculture_pin']

agri_data_all
```

```python
dict1 = dict(tuple(agri_data_all.groupby('element')))
agri_data = dict1['Gross Production 1999-2001 (1000 I$)']
agri_data.dropna(inplace=True)
# agri_data.drop(agri_data[agri_data['element'] != 'Gross Production 1999-2001 (1000 I$)'].index)
agri_data = agri_data[['country_or_area', 'year', 'value']]
agri_data.columns = ['Country', 'Year', 'Gross_Agri_Production']
# agri_data = agri_data[~agri_data.Country.str.endswith('+')]
agri_data
```

```python
agri_data_dict = dict(tuple(agri_data.groupby('Country')))
agri_data_country = agri_data_dict[country] #Select any country
agri_data_country
# agri_data_country.to_csv('test.csv')
```

```python
land_data = pd.read_csv('Dataset/archive/fao_data_land_data.csv') #country_or_area
land_data.drop(land_data[land_data['category'] != 'agricultural_area'].index, inplace = True)
land_data.dropna(inplace=True)
land_data = land_data[['country_or_area', 'year', 'value']]
land_data.columns = ['Country', 'Year', 'Area']
agri_land_data = land_data[~land_data['Country'].str.endswith('+')]
agri_land_dict = dict(tuple(agri_land_data.groupby('Country')))
agri_land_country = agri_land_dict[country]
agri_land_country
```

```python
world_gdp_data = pd.read_csv('Dataset/archive/world_country_gdp_usd.csv')
world_gdp_data['Country Name'] = world_gdp_data['Country Name'].replace('United States', 'United States of America')
world_gdp_data = world_gdp_data.drop('Country Code', axis=1)
world_gdp_data.columns = ['Country', 'Year', 'GDP', 'GDP_Per_Capita']
world_gdp_data.dropna(inplace=True)
world_gdp_dict = dict(tuple(world_gdp_data.groupby('Country')))
world_gdp_country = world_gdp_dict[country]
world_gdp_country
```

```python
#Agriculture Research Spending
#Year varies from 2000 to 2012  
#6084	Spending, total	million PPP (constant 2011 prices)
world_research_data = pd.read_csv('Dataset/archive/ASTI_Research_Spending_E_All_Data.csv', encoding='ISO-8859-1') 
world_research_data = world_research_data[['Country', 'Element', 'Year', 'Value']]
world_research_data.drop(world_research_data[world_research_data['Element'] != '2011 PPP$'].index, inplace = True)
world_research_data.drop('Element', axis=1, inplace=True)
world_research_data.columns = ['Country', 'Year', 'Agriculture Research Spending']
# world_gdp_data.dropna(inplace=True)
world_research_dict = dict(tuple(world_research_data.groupby('Country')))
world_research_country = world_research_dict[country]
world_research_country
#Inflation nos?
```

```python
# Year varies from 2002 to 2014
# Unit kg/ha
world_fertilizer_data = pd.read_csv('Dataset/archive/Environment_Fertilizers_E_All_Data.csv', encoding='ISO-8859-1') #Area
grouped_fertilizer_df = world_fertilizer_data.groupby(['Area', 'Year', 'Item Code'])['Value'].sum().reset_index()
pivot_df = grouped_fertilizer_df.pivot_table(index=['Area', 'Year'], columns='Item Code', values='Value', fill_value=0).reset_index()
pivot_df['Sum'] =  pivot_df[3102] + pivot_df[3103] + pivot_df[3104]
fertilizer_data = pivot_df[['Area', 'Year', 'Sum']]
fertilizer_data.columns = ['Country', 'Year', 'Total Fertilizer Use']
world_fertilizer_dict = dict(tuple(fertilizer_data.groupby('Country')))
world_fertilizer_country = world_fertilizer_dict[country]
world_fertilizer_country
```

```python
world_temp_data = pd.read_csv('Dataset/archive/Environment_Temperature_change_E_All_Data.csv', encoding='ISO-8859-1') #Area
print(world_temp_data['Months'].unique())
print(world_temp_data['Element'].unique())
world_temp_dict = dict(tuple(world_temp_data.groupby('Area')))
world_temp_country = world_temp_dict[country]
# world_temp_country.to_csv('temp-test.csv')
world_temp_change_country  = world_temp_country[(world_temp_country['Months'] == 'Meteorological year') & (world_temp_country['Element'] == 'Temperature change')]
world_temp_change_country = world_temp_change_country.rename(columns={'Value': 'Temp. Change', 'Area': 'Country' })
world_temp_change_country
# 6 degree celsius change in 10 years?
```

```python
world_temp_deviation_country  = world_temp_country[(world_temp_country['Months'] == 'Meteorological year') & (world_temp_country['Element'] == 'Standard Deviation')]
world_temp_deviation_country = world_temp_deviation_country.rename(columns={'Value': 'Temp. Deviation', 'Area': 'Country' })
world_temp_deviation_country
```

```python
#Merge dataframes
# Merge the DataFrames one by one based on 'Year'
merged_df = pd.merge(agri_data_country, agri_land_country, on='Year', how='inner')
merged_df = pd.merge(merged_df, world_gdp_country, on='Year', how='inner')
merged_df = pd.merge(merged_df, world_research_country, on='Year', how='inner')
merged_df = pd.merge(merged_df, world_fertilizer_country, on='Year', how='inner')
merged_df = pd.merge(merged_df, world_temp_change_country, on='Year', how='inner')
merged_df = pd.merge(merged_df, world_temp_deviation_country, on='Year', how='inner')
merged_df = merged_df[['Year', 'Gross_Agri_Production', 'Area', 'GDP', 'GDP_Per_Capita', 'Agriculture Research Spending', 'Total Fertilizer Use','Temp. Change', 'Temp. Deviation' ]]
merged_df
```

```python
Y_Prod = merged_df['Gross_Agri_Production']
# X_Prod = merged_df[['Year', 'Area', 'GDP', 'GDP_Per_Capita',
#        'Agriculture Research Spending', 'Total Fertilizer Use', 'Temp. Change']]

X_Prod = merged_df[['Year', 'Area', 'GDP', 'GDP_Per_Capita',
        'Agriculture Research Spending', 'Total Fertilizer Use']]

X_Prod = sm.add_constant(X_Prod)
model = sm.OLS(Y_Prod, X_Prod).fit()
print(model.summary())
```
